```python
import seaborn as sns
%matplotlib inline
```

```python
planets = sns.load_dataset('planets')
```

```python
planets.head()
```

```python
#displot - показує розподіл одновимірного набору спостережень
#kde - за замовчуванням True, якщо kde=False - то на графіку буде лише гістограма без кривої розподілу
#bins - ширина "кроку" для гістограми
sns.distplot(planets['year'], bins=10)
```

```python
sns.jointplot(x='year', y='number', data = planets, kind="resid", color="g")
```

```python
sns.pairplot(planets)
```

```python
import statsmodels.nonparametric.api as smnp
sns.pairplot(planets,hue='number',palette='GnBu_d')
```

```python
sns.countplot(x='number', data=planets)
```

```python
sns.boxplot(x="number",y="year",data=planets,palette="rainbow")
```

```python
sns.stripplot(x = 'number', y = 'distance', data = planets, size = 12, palette = 'GnBu_d' )
```

```python
sns.heatmap(planets.corr())
```

```python
from matplotlib import pyplot as plt
g = sns.FacetGrid(planets, col='number', row='method')
g = g.map(plt.hist, 'year')
```
